```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# create a plot with ax.plt() (MATLAB METHOD)
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
fig.show()
```

```python
# alternative with ax.plt() (MATLAB METHOD)
# to change figure-level attributes (such as axis labels, title, legend, etc.) or save the figure as as image use fig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
fig.show()
```

```python
# Aaaand another way to create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Add a grid
plt.grid()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
# make the yaxis go from Jan-Dec rather than Dec-Jan
plt.gca().invert_yaxis()
# set axis labels
plt.xlabel = "Fare($)"
plt.ylabel = "Date"
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
fig.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label = 'Chicago', color = "cyan")
ax.set_xlabel('Fare($)')
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.legend()
```
