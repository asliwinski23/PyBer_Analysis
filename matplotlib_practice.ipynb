```python
import matplotlib
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# alternative with ax.plt() (MATLAB METHOD)
# to change figure-level attributes (such as axis labels, title, legend, etc.) or save the figure as as image use fig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
fig.show()
```

```python
ml = MultipleLocator(1)

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Add a grid
plt.grid()
# adjust y limits
plt.ylim(-5,60)
plt.yticks(np.arange(0, 61, step=5.0))
# get the standard deviation for the values in the y_axis
stdev = statistics.stdev(y_axis)
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize = 5)
# Create minor ticks at an increment of 1.
# plt.minorticks_on() -- turns ticks on for both axis --> default is 1
plt.axes().yaxis.set_minor_locator(ml)
plt.show()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
# make the yaxis go from Jan-Dec rather than Dec-Jan
plt.gca().invert_yaxis()
# set axis labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
fig.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label = 'Chicago', color = "cyan")
ax.set_xlabel('Fare($)')
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.legend()
```

```python
plt.plot(y_axis,x_axis,'o', color = "red")
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
# if have multiple datasets in one plot, can make a list of legend titles in square brackets
plt.legend(["Chicago"])
plt.show()
```

```python
plt.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], c = "skyblue", alpha = 0.5, edgecolor = "b", linewidth = 2)
plt.legend(["Boston"])
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.ylabel("Fare ($)")
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct='%.1f%%', colors = colors, startangle = 90, counterclock = False)
plt.show()
```
